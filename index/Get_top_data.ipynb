```python
#! Export top tickers

#* Open files
with open('Indici_top.txt', 'r') as f:
    cod = [int(line) for line in f.read().splitlines()]

with open('Tickers.txt', 'r') as file:
    names = file.readlines()

names = [name.strip() for name in names]

#* Extract top tickers
i = 0
top_tickers = []
for codice in cod:
    top_tickers.append(names[codice])
    i += 1

#* Export sylmbols
with open('Tickers_top.txt', 'w') as file:
    for symbol in top_tickers:
        file.write(f'{symbol}\n')
```

```python
#! Download historical data from Yahoo Finance

# Import libraries
import yfinance as yf
import pandas as pd
import os
import csv

# Carica il file CSV
with open('Tickers_top.txt', 'r') as file:
    symbols_list = [line.strip() for line in file]

# Download data
data = pd.DataFrame()


# Scarica tutti i dati
all_data = yf.download(symbols_list, start="2022-01-01", end="2023-01-01", period="1d")

# Seleziona le colonne 'Close' e 'Volume'
group_data = all_data[["Close", "Volume"]]


data = pd.concat([data, group_data], axis=1)

# Clear and export data
data_cleaned = data.drop(data.columns[data.isna().any()].tolist(), axis=1)
data_cleaned = data_cleaned.round(4)

print(type(data_cleaned), data_cleaned.shape, data_cleaned)

data_cleaned.to_csv('Historical_data_top', index=False)
```
